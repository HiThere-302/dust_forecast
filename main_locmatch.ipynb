- 위도와 경도를 좌표로 생각한다.
- 두 좌표의 거리가 가장 가까운 지역의 날씨 정보에 영향을 받는다.
- x = Latitude, y = Longitude
- distance = sqrt((x_1 - x_2)^2 + (y_1-y_2)^2)

```python
awsmap = pd.read_csv('META/awsmap.csv')
pmmap = pd.read_csv('META/pmmap.csv')

aws_lat = np.array(awsmap['Latitude']) # 위도 
aws_long = np.array(awsmap['Longitude']) # 경도 
aws_loc = np.array(awsmap['Location']) # 탐지소 지역명 
```

```python
distance_lat = list()
distance_long = list()
for index, row in pmmap.iterrows():
    distance_lat.append(aws_lat - row[1])
    distance_long.append(aws_long - row[2])
distance_lat = np.array(distance_lat)
distnace_long = np.array(distance_long)

distance = np.sqrt(np.square(distance_lat) + np.square(distance_long))

# 가장 가까운 거리
min_loc = list()
for d in distance:
    min_loc.append(np.argmin(d))

# 그곳의 이름
loc_match = dict()
for index, row in pmmap.iterrows():
    loc_match[row[0]] = aws_loc[min_loc[index]]
    
# 가나다순 정리
loc_match = dict(sorted(loc_match.items()))
```
